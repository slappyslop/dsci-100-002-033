```r
library(tidyverse)
housing <- read_csv("data/HouseListings-Top45Cities-10292023-kaggle.csv")
housing
```

```r
We want to find the average costs per city.
```

```r
table_avg_cost <- housing |>
select(City, Price) |>
group_by(City) |>
summarize(avg_cost = mean(Price)) |>
arrange(-avg_cost)
# pivot_wider(names_from = City, values_from = avg_cost) 
table_avg_cost
```

```r
By arranging the average costs in descending order, we can compare them with the median family income and the number of beds in each of the cities
```
